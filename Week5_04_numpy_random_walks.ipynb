```python
import random

position = 0
walk = [ position ]#그때 그때 position에 대해 기록할 list
steps = 1000 #1000보 걸음

for i in range(steps):
    step = 1 if random.randint(0,1) else -1
    position += step
    walk.append(position)

print(walk)
```

```python
#numpy를 통해 똑같은걸 간략하게 나타낼 수 있다...

import numpy as np
nsteps = 1000
draws = np.random.randint(0, 2, size=nsteps)
steps = np.where(draws > 0 , 1, -1) #draws > 0 이면 1, <=0 이면 -1로 추정
walk = steps.cumsum()

walk
```

```python
#좀 더 판을 키움

nwalks = 5000
nsteps = 1000

draws = np.random.randint(0, 2, size=(nwalks, nsteps)) # 0 or 1로 난수 생성, 크기는 nwalks x nsteps
steps = np.where(draws>0, 1, -1) #1이면 steps=1, 0이면 steps=-1
walks = steps.cumsum(1) #steps의 누적합..???????인자 1은 뭐지

walks
```

```python
hits30 = (np.abs(walks) >= 30).any(1) 
hits30
```

```python
hits30.sum() #number that hit 30 or -30
```

```python
crossing_times = (np.abs(walks[hits30])>=30).argmax(1)
crossing_times
```

```python
crossing_times.mean()
```
